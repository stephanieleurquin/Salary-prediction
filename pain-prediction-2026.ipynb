```python

import pandas as pd
import matplotlib.pyplot as plt
import os #pour gérer les chemins de fichier

# ✅ Chemin exact du CSV sur Kaggle
chemin_csv = "/kaggle/input/datasets/devashishmani1/pain-prediction-dataset-emphatica-e4/pain_dataset_200P_4hz.csv"

# Vérifier que le fichier existe
if not os.path.exists(chemin_csv):
    raise FileNotFoundError(f"❌ Fichier introuvable : {chemin_csv}")
else:
    print("✅ Fichier trouvé !")

# Charger le CSV
df = pd.read_csv(chemin_csv)
print("✅ CSV chargé !\n")
print("--- Aperçu du dataset ---")
print(df.head())

# Vérifier la colonne 'pain_scale'
if 'pain_scale' in df.columns:
    # Convertir en numérique (pour éviter les erreurs)
    df['pain_scale'] = pd.to_numeric(df['pain_scale'], errors='coerce')
    
    # Graphe des 500 premières mesures
    plt.figure(figsize=(12,4))
    plt.plot(df['pain_scale'][:500], marker='o')
    plt.title("Pain Scale (500 premières mesures)")
    plt.xlabel("Index")
    plt.ylabel("Pain Scale")
    plt.grid(True)
    plt.show()
    
    # Histogramme complet
    plt.figure(figsize=(12,5))
    plt.hist(df['pain_scale'].dropna(), bins=20, color='skyblue', edgecolor='black')
    plt.title("Distribution complète de Pain Scale")
    plt.xlabel("Pain Scale")
    plt.ylabel("Nombre de mesures")
    plt.show()
    
    # Statistiques descriptives
    print("\n--- Statistiques sur pain_scale ---")
    print(df['pain_scale'].describe())
else:
    print("❌ La colonne 'pain_scale' n'existe pas dans le dataset")
```
